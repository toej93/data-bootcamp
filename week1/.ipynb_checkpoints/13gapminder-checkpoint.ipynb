## Load some data



You may have thought that the argument to `read_csv` needed to be a
file, but it can also be a URL.  Let&rsquo;s load some gapminder data from a
GitHub repository.



```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
url = 'https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv'
df = pd.read_csv(url)
df.tail(5)
```

## Some initial plots



We could explore GDP per capita versus life expectancy.



```python
df.plot.scatter( 'gdpPercap', 'lifeExp' )
plt.show()
```

It might be a bit easier to see a football-shaped scatterplot if we
use a logarithmic scale on the $x$-axis.



```python
df.plot.scatter( 'gdpPercap', 'lifeExp' )
plt.xscale('log')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import BayesianRidge
from sklearn.linear_model import Ridge
from sklearn.linear_model import LogisticRegression

lm = LinearRegression().fit( df[['gdpPercap']], df['lifeExp'] )
br = BayesianRidge().fit( df[['gdpPercap']], df['lifeExp'] )
ls = Ridge(alpha=.5).fit( df[['gdpPercap']], df['lifeExp'] )
#logreg = LogisticRegression(solver='lbfgs', multi_class='auto').fit( df[['gdpPercap']], df['lifeExp'] )
```

```python
df.plot.scatter( 'gdpPercap', 'lifeExp' )
plt.plot( df['gdpPercap'], ls.predict(df[['gdpPercap']]), color='red', linewidth=2 )
#plt.plot( df['gdpPercap'], ls.predict(df[['gdpPercap']]), color='blue', linewidth=2 )
plt.xscale('log')
plt.show()
```

```python
from sklearn.model_selection import KFold
model = BayesianRidge()
scores = []

X = df[['gdpPercap']]
y = df['lifeExp']

kf = KFold(n_splits=5, shuffle=True, random_state=1)
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    model.fit(X_train, y_train)
    score = model.score(X_test, y_test)
    scores.append(score)

scores
```
